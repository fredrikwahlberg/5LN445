```
import numpy as np
from numpy.linalg import norm

import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.decomposition import PCA
```

# Subspace projection

```
n_clusters = 3

# Setup data and plotting
data, labels = make_blobs(n_samples=n_clusters*10, n_features=3, centers=n_clusters)
fig = plt.figure(figsize=(11, 5), dpi=100)
# Plot 3D
ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.scatter(data[:, 0], data[:, 1], data[:, 2], s=50, c=labels, marker='o', cmap='jet', edgecolors='k')
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
# Plot 2D
ax = fig.add_subplot(1, 2, 2)
subspace_data = PCA(n_components=2).fit_transform(data)
ax.scatter(subspace_data[:, 0], subspace_data[:, 1], s=50, c=labels, marker='o', cmap='jet', edgecolors='k')
# Format and commit to screen
fig.tight_layout()
fig.show()
```

```
model = PCA(n_components=2)
subspace_data = model.fit_transform(data)
# subspace_data
np.matmul(data, model.components_.T)
```
